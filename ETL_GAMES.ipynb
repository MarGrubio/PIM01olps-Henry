### **Transformación de la columnas "price", "release_date" y "genres"**

Verificamos que los datos de la columna price sea númericos y si no lo son revisamos cuales deben ser reemplazados 

```python
Games['price'].unique()
```

Se crea una función para poder reemplazar los datos de tipo str por tipo float 0.0

```python
# Definir una función lambda que convierte el valor a float o devuelve 0.0 en caso de error
cambio_a_float = lambda valor: float(valor) if isinstance(valor, (float, int)) else 0.0 if pd.isna(valor) else 0.0

# Aplicar la función lambda a la columna precio
SteamGames["price"] = SteamGames["price"].apply(cambio_a_float)
SteamGames.info()
```

```python
SteamGames["price"].value_counts()
```

Se buscan outliers en precio, se identifican los umbrales y se identifican los outliers 

```python
price_description = SteamGames['price'].describe()
Q1 = price_description['25%']
Q3 = price_description['75%']
IQR = Q3 - Q1

print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)
```

```python
Superior = SteamGames['price'].quantile(0.75) + 1.5 * (SteamGames['price'].quantile(0.75) - SteamGames['price'].quantile(0.25))
Inferior = SteamGames['price'].quantile(0.25) - 1.5 * (SteamGames['price'].quantile(0.75) - SteamGames['price'].quantile(0.25))

print("Umbral Superior:", Superior)
print("Umbral Inferior:", Inferior)
```

```python
outliers_sup = SteamGames.loc[SteamGames['price'] > Superior]
outliers_inf = SteamGames.loc[SteamGames['price'] < Inferior]

print(outliers_sup)
print(outliers_inf)
```

```python
# Convierte la columna 'release_date' a cadenas de texto (strings)
# Define una función lambda para extraer el año o asignar 'sin dato' en caso de que no haya un año válido
# Aplica la función lambda a la columna 'release_date' y asigna el resultado a 'Año_estreno'

SteamGames['release_date'] = SteamGames['release_date'].astype(str)
extract_year = lambda date: re.search(r'\d{4}', date).group() if re.search(r'\d{4}', date) else 'sin dato'
SteamGames['Año_estreno'] = SteamGames['release_date'].apply(extract_year)
```
