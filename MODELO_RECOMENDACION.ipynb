Para el modelamiento, vamos a utilizar una librería que incluye una herramienta llamada 'Similitud de Coseno'. Esta herramienta nos permite medir cuán parecidas son dos variables, y los valores de similitud pueden variar entre -1 y 1. Cuando la similitud es 1, significa que las variables son muy parecidas, mientras que un valor cercano a 0 indica poca similitud, y -1 señala que son totalmente opuestas.

En nuestro caso, solo vamos a utilizar tres columnas específicas de nuestro DataFrame:

3. 'genre': Esta columna representa el género al que pertenece cada juego.

Estas tres columnas son las que utilizaremos para calcular la similitud de coseno y realizar nuestro modelamiento.
